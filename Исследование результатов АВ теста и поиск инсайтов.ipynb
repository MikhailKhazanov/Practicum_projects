SollmaFin — международное мобильное приложение для онлайн-торговли финансовыми активами. Пользователи могут инвестировать деньги в акции, валюту, криптовалюту, биржевые фонды (ETF) и другие активы. Целевая аудитория приложения — начинающие инвесторы. Получить доступ к финансовым рынкам можно со смартфона либо через веб-версию. 

Приложение ориентировано на рынок Латинской Америки с акцентом на четыре страны: Мексику, Бразилию, Колумбию и Аргентину — там быстро растёт интерес к финтеху и инвестициям, а экономики очень разные. 

Появилась гипотеза о недостаточной финансовой грамотности пользователей: они не понимают, как работать с активами разной степени риска. Клиенты покупают активы с высоким риском и теряют деньги, что снижает вовлечённость.
Команда продукта решила обновить онбординг пользователей и добавить в него детальную информацию о различиях финансовых активов и связанных с ними рисках. Однако появилось опасение, что углублённый онбординг отпугнёт клиентов от пополнения депозита и более рискованных вложений. 

Чтобы разобраться, команда решила провести A/B-эксперимент. 

### А/В-эксперимент

В эксперименте участвовали новые пользователи, которые зарегистрировались в приложении со 2 по 15 июня 2025 года. 
Пользователей случайным образом разделили на две равные группы: 
- Контрольная группа проходила стандартный онбординг без обязательного обучения.
- Тестовая группа проходила обновлённый онбординг с подробной информацией об активах и связанных с ними рисках. 

После разделения активность пользователей анализировали в течение недели.

Цель эксперимента — оценить, как обновлённый онбординг влияет на поведение пользователей и их дальнейшую инвестиционную активность.

При внедрении новой фичи команда продукта выдвинула такие гипотезы:
- Гипотеза роста: обучающий онбординг помогает пользователям лучше понимать принципы инвестирования, поэтому они будут чаще открывать второй депозит.
- Гипотеза риска: информация о возможных потерях и высоких рисках отпугнёт некоторых новичков, особенно самых осторожных, что снизит конверсию в первый депозит.
- Дополнительная гипотеза: после нового онбординга пользователи, которые выбрали высокорискованные активы, будут чаще, чем раньше, возвращаться и открывать второй депозит. При старом онбординге пользователи часто покупали активы с высоким риском без понимания последствий. Это приводило к потерям и оттоку после первого депозита. 

Чтобы увидеть полную картину, недостаточно следить за метриками конверсии. Команда выбрала для эксперимента такой набор метрик: 
- Ключевая метрика — средняя сумма всех депозитов на одного пользователя (включая тех, кто установил приложение или открыл веб-версию).
- Барьерная метрика — конверсия из регистрации в первый депозит.
- Вспомогательная метрика 1 — конверсия из первого депозита во второй.
- Вспомогательная метрика 2 — средняя сумма всех депозитов на пользователя, который открыл хотя бы один депозит.

Ожидалось, что ключевая и барьерная метрики не упадут, а вспомогательные покажут значительный рост.

**Задача 2.2.** Анализ воронок событий.
- Постройте классическую и пошаговую воронки;
- Охарактеризуйте пользовательский путь в приложении и определите узкие места воронки. 


````python
# Задаем порядок шагов в воронке
funnel_steps = ['install / open_web', 'introduction', 'registration', 'main_page',
 'onboarding_complete', 'first_deposit', 'asset_purchase', 'second_deposit']


# Фильтруем датасет оставляя только нужные шаги воронки
funnel_df_steps = df[df['event_name'].isin(funnel_steps)]
# Считаем число уникальных пользователей на каждом шаге
funnel_counts = funnel_df_steps.groupby('event_name')['user_id'].nunique().reindex(funnel_steps)


# Собираем таблицу воронки
funnel_df = pd.DataFrame({
   'step': funnel_steps,
   'users': funnel_counts.values
})

# Добавляем конверсии и округляем до 1 знака после запятой
funnel_df['conversion_from_first_%'] = round((funnel_df['users'] / funnel_df['users'].iloc[0] * 100), 1)
funnel_df['conversion_from_previous_%'] = round((funnel_df['users'] / funnel_df['users'].shift(1) * 100), 1)

display(funnel_df.head(10))
````

````python
# Визуализируем данные, кроме первого шага, чтобы не выводить NaN
plt.figure(figsize=(10, 6))
# Строим barplot

plt.bar(funnel_df['step'], funnel_df['conversion_from_first_%'], color='skyblue')
plt.xticks(funnel_df['step'], rotation=45, ha='right')
plt.title('Классическая воронка: конверсия от первого шага')
plt.ylabel('Конверсия от первого шага (%)')
plt.xlabel('Этапы воронки')
plt.show()
````

На классической воронке падение начинается после онбординга на переходе к первому депозиту (87% → 67% пользователей от первого шага), от первого депозита к покупке активов (67% → 37% пользователей от первого шага), от покупки активов к второму депозиту (37% → 13% ользователей от первого шага).

````python
# Визуализируем данные, кроме первого шага, чтобы не выводить NaN
plt.figure(figsize=(10, 6))
# Строим barplot

plt.bar(funnel_df['step'], funnel_df['conversion_from_previous_%'], color='skyblue')
plt.xticks(funnel_df['step'], rotation=45, ha='right')
plt.title('Step-by-step воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.show()
````

На пошаговой воронке падение начинается после онбординга на переходе к первому депозиту (98% → 80% пользователей от предыдущего шага), от первого депозита к покупке активов (80% → 55% пользователей от предыдущего шага), от покупки активов к второму депозиту (55% → 35% пользователей от предыдущего шага).

**Задача 2.3.** Влияние уровня риска актива на открытие второго депозита.
- Изучите, как уровень риска первого актива влияет на конверсию во второй депозит.  

````python
# Пользователи с первым депозитом
first_deposit_users = df[df['event_name'] == 'first_deposit']['user_id'].unique()
users_with_deposit = df[df['user_id'].isin(first_deposit_users)].copy()

# Определение первого купленного актива
first_asset_purchase = users_with_deposit[users_with_deposit['event_name'] == 'asset_purchase'].sort_values('event_ts').drop_duplicates('user_id', keep='first')

# Пользователи со вторым депозитом
second_deposit_users = users_with_deposit[users_with_deposit['event_name'] == 'second_deposit']['user_id'].unique()

# Объединение информации о первом и втором депозитах
first_asset_purchase['made_second_deposit'] = first_asset_purchase['user_id'].isin(second_deposit_users)

# Группируем по уровню риска и считаем конверсию 
risk_conversion = first_asset_purchase.groupby('risk_level').agg(
    total_users=('user_id', 'nunique'),
    users_with_second_deposit=('made_second_deposit', 'sum')
).reset_index()

risk_conversion['conversion_to_second_deposit_%'] = (
    risk_conversion['users_with_second_deposit'] / risk_conversion['total_users'] * 100
).round(1)

print(risk_conversion)
````

Ожидаемо, что самая низкая конверсия (26.4%) на высоком уровне риска, так как клиенты с высокой вероятностью могут потерять средства. На низком уровне конверсия 46.4%, на среднем - 43%.

**Задача 2.4.** Анализ ключевой метрики на исторических данных.
- На исторических данных рассчитайте значение ключевой метрики — средней суммы всех депозитов на одного пользователя. Учитывайте всех пользователей, которые взаимодействовали с сервисом. Это позволит зафиксировать базовый уровень монетизации.
- Визуализируйте данные, где необходимо. По результатам исследовательского анализа сделайте краткий вывод. 

````python
# общая сумма депозитов по всем пользователям
total_deposits = df['amount'].sum()

# общее количество уникальных пользователей
total_users = df['user_id'].nunique()

# ARPU 
ARPU = total_deposits / total_users

# общая сумма депозитов для пользователя
deposits_by_user = df.groupby('user_id')['amount'].sum().reset_index()

# Объединение всех пользователей, в том числе без депозитов
all_users_deposits = df['user_id'].unique()
deposits_series = deposits_by_user.set_index('user_id')['amount'].reindex(all_users_deposits, fill_value=0)

print(f"Общая сумма всех депозитов: {total_deposits}")
print(f"Общее количество уникальных пользователей: {total_users}")
print(f"Средняя сумма депозитов на пользователя (ARPU): {ARPU:.2f}")
print(f"Стандартное отклонение суммы депозитов: {deposits_series.std():.2f}")
print(f"• Распределение скошено вправо с длинным хвостом - большинство пользователей вносят небольшие суммы")
````

````python
plt.figure(figsize=(14, 6))
sns.histplot(deposits_series, bins=50, kde=True)
plt.title('Распределение суммы депозитов на пользователя', fontsize=16)
plt.xlabel('Сумма депозитов', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)
plt.grid(True)
plt.show()
````

````python
# Фильтрация выбросов по 99 перцентилю
percentile99 = deposits_series.quantile(0.99)
filtered_deposits = deposits_series[deposits_series <= percentile99]

plt.figure(figsize=(14, 6))
sns.histplot(filtered_deposits, bins=50, kde=True)
plt.title(f'Распределение суммы депозитов на пользователя', fontsize=16)
plt.xlabel('Сумма депозитов', fontsize=12)
plt.ylabel('Количество пользователей', fontsize=12)
plt.grid(True)
plt.show()
````

Если посчитать среднюю сумму депозитов на всех пользователей, то получается 91.32 доллара США. При этом стандартное отклонение составило 110.84, это примерно в 1.2 раза больше. График скошен сильно вправо, отфильтровали по 99 перцентилю крайние значения свыше 520 долларов США.

- Сравните результаты и оцените, есть ли статистически значимая разница между группами. 

````python
# Пользователи, открывшие депозит и сделавшие покупку
users_with_asset_purchase = df_abt[df_abt['event_name'] == 'asset_purchase'].copy()

# Пользователи со вторым депозитом
users_with_second_deposit = df_abt[df_abt['event_name'] == 'second_deposit']['user_id'].unique()

# Добавим информацию о втором депозите 
users_with_asset_purchase['made_second_deposit'] = users_with_asset_purchase['user_id'].isin(users_with_second_deposit)

# Добавляем уровень риска
risk_conversion_group = users_with_asset_purchase.groupby(['group', 'risk_level']).agg(
    total_users=('user_id', 'nunique'), 
    users_with_second_deposit=('made_second_deposit', 'sum')).reset_index()

# Конверсия
risk_conversion_group['conversion_to_second_deposit_%'] = (risk_conversion_group['users_with_second_deposit'] / risk_conversion_group['total_users'] * 100).round(1)

display(risk_conversion_group)
````

````python
# Используем z-тест пропорций
def z_test(group_a, group_b):
    stat, p_value = proportions_ztest(
        count=[group_a['users_with_second_deposit'].iloc[0], group_b['users_with_second_deposit'].iloc[0]],
        nobs=[group_a['total_users'].iloc[0], group_b['total_users'].iloc[0]]
    )
    return stat, p_value

results = []
for risk_level in ['high', 'medium', 'low']:
    control = risk_conversion_group[(risk_conversion_group['group'] == 'control') & (risk_conversion_group['risk_level'] == risk_level)]
    test = risk_conversion_group[(risk_conversion_group['group'] == 'test') & (risk_conversion_group['risk_level'] == risk_level)]
    if control.empty or test.empty:
        p_value = np.nan
    else:
        print('Н0: Конверсия во второй депозит в тестовой группе не отличается от конверсии контрольной группы')
        print('Н1: Конверсия во второй депозит в тестовой группе отличается от конверсии контрольной группы')
        
        stat, p_value = z_test(control, test)
        
    print(f"p-value: {p_value:.4f}")
    if p_value < 0.05:
        print("Вывод: p-value < 0.05. Разница статистически значима. Отвергаем H0.")
    else:
        print("Вывод: p-value >= 0.05. Разница статистически не значима. Не отвергаем Н0.")
    
    results.append({
    'risk_level': risk_level,
    'p_value': p_value,
    })
    
results_df = pd.DataFrame(results)
````

В группе с high риском конверсия после внедрения фичи резко увеличилась (46% против 27.7%), p-value составляет 0.0000, альтернативную гипотезу принимаем.

В группе medium конверсия стала 47% против 44.5% в контрольной группе. P-value составил 0.3609, значимого эффекта нет, оставляем нулевую гипотезу.

В группе low конверсия стала 57.5% против 49.8% в контрольной группе. P-value составляет 0.0391, альтернативную гипотезу принимаем.

Сделайте промежуточный вывод. 


````python
# Получаем всех пользователей по группам
all_users = df_abt[['user_id', 'group']].drop_duplicates()

# 1. Ключевая метрика: средняя сумма всех депозитов на одного пользователя
deposit_events = df_abt[df_abt['event_name'].isin(['first_deposit', 'second_deposit'])].copy()

user_total_deposits = deposit_events.groupby(['user_id', 'group'])['amount'].sum().reset_index()
all_users_with_deposits = all_users.merge(user_total_deposits, on=['user_id', 'group'], how='left')
all_users_with_deposits['amount'] = all_users_with_deposits['amount'].fillna(0)

# 2. Барьерная метрика: конверсия из регистрации в первый депозит
registrations = df_abt[df_abt['event_name'] == 'registration'][['user_id', 'group']]
first_deposit_users = df_abt[df_abt['event_name'] == 'first_deposit'][['user_id', 'group']]

# 3. Вспомогательная метрика 1: конверсия из первого депозита во второй
second_deposit_users = df_abt[df_abt['event_name'] == 'second_deposit'][['user_id', 'group']].drop_duplicates()

# 4. Вспомогательная метрика 2: средняя сумма всех депозитов на пользователя с хотя бы одним депозитом
users_with_any_deposit = deposit_events[['user_id', 'group', 'amount']]

````

````python
# Расчет метрик по группам
def calculate_metrics():
    results = []
    
    for group_name in ['control', 'test']:
        # Фильтруем данные по группе
        group_all_users = all_users[all_users['group'] == group_name]
        group_users_with_deposits = all_users_with_deposits[all_users_with_deposits['group'] == group_name]
        group_registrations = registrations[registrations['group'] == group_name]
        group_first_deposit = first_deposit_users[first_deposit_users['group'] == group_name]
        group_second_deposit = second_deposit_users[second_deposit_users['group'] == group_name]
        group_users_with_any_deposit = users_with_any_deposit[users_with_any_deposit['group'] == group_name]
        
        # Ключевая метрика
        avg_deposit_per_user = group_users_with_deposits['amount'].mean()
        
        # Барьерная метрика
        reg_conversion = (len(group_first_deposit) / len(group_registrations) * 100) if len(group_registrations) > 0 else 0
        
        # Вспомогательная метрика 1
        second_deposit_conversion = (len(group_second_deposit) / len(group_first_deposit) * 100) if len(group_first_deposit) > 0 else 0
        
        # Вспомогательная метрика 2
        if len(group_users_with_any_deposit) > 0:
            avg_deposit_for_active = group_users_with_any_deposit.groupby('user_id')['amount'].sum().mean()
        else:
            avg_deposit_for_active = 0
        
        results.append({
            'group': group_name,
            'avg_deposit_per_user': avg_deposit_per_user,
            'registration_conversion_%': round(reg_conversion, 2),
            'second_deposit_conversion_%': round(second_deposit_conversion, 2),
            'avg_deposit_for_active': avg_deposit_for_active
        })
    
    return pd.DataFrame(results)

# Рассчитываем метрики
metrics_df = calculate_metrics()
print("\nМетрики по группам:")
print(metrics_df.to_string(index=False))

# Получаем значения для удобства
control_metrics = metrics_df[metrics_df['group'] == 'control'].iloc[0]
test_metrics = metrics_df[metrics_df['group'] == 'test'].iloc[0]

````

````python
# Анализ динамики ключевой метрики (ARPU)
# Создаем временные колонки для группировки по дате
ab_data_full = df_abt.copy()
ab_data_full['event_ts_date'] = pd.to_datetime(ab_data_full['event_ts']).dt.date
ab_data_full['first_dt_date'] = pd.to_datetime(ab_data_full['first_dt']).dt.date

# Рассчитываем ежедневное количество новых пользователей по группам
daily_new_users = ab_data_full.groupby(['first_dt_date', 'group'])['user_id'].nunique().reset_index()
daily_new_users.columns = ['date', 'group', 'new_users']
daily_new_users['date'] = pd.to_datetime(daily_new_users['date'])

# Рассчитываем ежедневную сумму депозитов по группам
daily_deposits = ab_data_full.groupby(['event_ts_date', 'group'])['amount'].sum().reset_index()
daily_deposits.columns = ['date', 'group', 'daily_deposits']
daily_deposits['date'] = pd.to_datetime(daily_deposits['date'])

# Объединяем данные и рассчитываем кумулятивные метрики
all_dates = pd.date_range(start=ab_data_full['first_dt_date'].min(), end=ab_data_full['event_ts_date'].max()).date
all_groups = ab_data_full['group'].unique()
full_index = pd.MultiIndex.from_product([all_dates, all_groups], names=['date', 'group'])

# Объединяем и заполняем пропущенные даты нулями
daily_data = pd.DataFrame(index=full_index).reset_index()
daily_data['date'] = pd.to_datetime(daily_data['date'])
daily_data = daily_data.merge(daily_new_users, on=['date', 'group'], how='left')
daily_data = daily_data.merge(daily_deposits, on=['date', 'group'], how='left')
daily_data.fillna(0, inplace=True)

# Рассчитываем кумулятивные значения
daily_data['cumulative_users'] = daily_data.groupby('group')['new_users'].cumsum()
daily_data['cumulative_deposits'] = daily_data.groupby('group')['daily_deposits'].cumsum()
# Расчет кумулятивной ARPU
daily_data['arpu_cum'] = daily_data['cumulative_deposits'] / daily_data['cumulative_users']

# Вывод данных для расчета кумулятивной ARPU
print("Данные для графика накопленной ARPU:")
display(daily_data[['date', 'group', 'arpu_cum']])
````

````python
# Бутстрап анализ
def bootstrap_percentiles(control_data, test_data, n_bootstrap=10000):
    np.random.seed(42)
    
    # Статистики для каждого бутстрапа
    bootstrap_results = []
    
    for i in range(n_bootstrap):
        # Бутстрап выборки
        control_sample = np.random.choice(control_data, size=len(control_data), replace=True)
        test_sample = np.random.choice(test_data, size=len(test_data), replace=True)
        
        # Рассчитываем перцентили
        control_percentiles = np.percentile(control_sample, [25, 50, 75])
        test_percentiles = np.percentile(test_sample, [25, 50, 75])
        
        # Рассчитываем разницы
        differences = test_percentiles - control_percentiles
        
        bootstrap_results.append(differences)
    
    return np.array(bootstrap_results)

````

````python
1. # Формулируем гипотезы

# H₀: Перцентиль (25 / 50 / 75) суммы депозитов одинаков между группами.
# H₁: Перцентиль суммы депозитов различается между тестом и контролем.

# Отбираем только события депозитов
deposits_abt = df_abt[df_abt['event_name'].isin(['first_deposit', 'second_deposit'])]

# Считаем общую сумму депозитов на одного пользователя
user_sums = deposits_abt.groupby('user_id')['amount'].sum().reset_index()

# Добавляем группу (test/control)
user_groups = df_abt[['user_id', 'group']].drop_duplicates()
user_sums = user_sums.merge(user_groups, on='user_id', how='left')


# Отбираем только пользователей с депозитами
user_sums = user_sums[user_sums['amount'] > 0]

# Делим на контрольную и тестовую группы
control = user_sums[user_sums['group'] == 'control']['amount'].values
test = user_sums[user_sums['group'] == 'test']['amount'].values


2. # Бустрап функция для расчетов по процентилям
def bootstrap_percentile_diff(control, test, percentile, n_iter=10000):
    np.random.seed(42)
    diffs = []
    for _ in range(n_iter):
        sample_c = np.random.choice(control, size=len(control), replace=True)
        sample_t = np.random.choice(test, size=len(test), replace=True)
        diff = np.percentile(sample_t, percentile) - np.percentile(sample_c, percentile)
        diffs.append(diff)
    ci = np.percentile(diffs, [2.5, 97.5])
    return diffs, ci

3.     # Расчеты для 25, 50, 75 перцентилей
results = {}
for p in [25, 50, 75]:
    diffs, ci = bootstrap_percentile_diff(control, test, percentile=p)
    results[p] = {
        'diffs': diffs,
        'ci': ci,
        'control_value': np.percentile(control, p),
        'test_value': np.percentile(test, p),
        'diff': np.percentile(test, p) - np.percentile(control, p)
    }



4.     # Визуализация 

plt.figure(figsize=(15, 4))
for i, p in enumerate([25, 50, 75]):
    plt.subplot(1, 3, i+1)
    plt.hist(results[p]['diffs'], bins=30, color='skyblue', edgecolor='black')
    plt.axvline(results[p]['ci'][0], color='red', linestyle='--', label='CI lower')
    plt.axvline(results[p]['ci'][1], color='red', linestyle='--', label='CI upper')
    plt.axvline(0, color='black', linestyle=':')
    plt.title(f'Разность {p}-го перцентиля')
    plt.xlabel('Тест - Контроль')
    plt.ylabel('Частота')
    plt.legend()

plt.tight_layout()
plt.show()
````
